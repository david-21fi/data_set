## Preprocesado

```python
dummy_sex = pd.get_dummies(data_set["Sex"],prefix="Sex")
data_set.drop("Sex",axis=1,inplace=True)
data_set = pd.concat([data_set,dummy_sex],axis=1)
data_set
```

# modelo de regresion

```python
colnames = data_set.columns.values.tolist()
colnames.remove("Age")
#colnames.remove("Sex_M")
#colnames.remove("Weight")
predictors = colnames
target = "Age"
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(data_set[predictors],data_set[target],test_size=0.25,random_state=1)
```

```python
lm = LinearRegression(n_jobs=2)
```

```python
lm.fit(x_train,y_train)
```

```python
lm.score(x_train,y_train)
```

```python
predicion = lm.predict(x_test)
list(zip(y_test,predicion))
```

```python
lm.score(x_test,y_test)
```

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVR
```

```python
estimador = SVR(kernel="linear")
selector = RFE(estimador,6,step=1)   
selector = selector.fit(data_set[predictors],data_set[target])
```

```python
list(zip(predictors,selector.support_,selector.ranking_))
```

# Utilizando random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor(n_estimators=500,oob_score=True,n_jobs=2)
```

```python
forest.fit(data_set[predictors],data_set[target])
```

```python
prediccion = forest.oob_prediction_

data_set["Prediccion_Forest"] = prediccion
data_set[[target,"Prediccion_Forest"]]
```

```python
forest.oob_score_
```

```python
data_set["Error_cuad"] = (data_set["Prediccion_Forest"]-data_set["Age"])**2
sum(data_set["Error_cuad"]/len(data_set))
```
