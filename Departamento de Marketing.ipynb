```python
data_df.loc[data_df["MINIMUM_PAYMENTS"].isnull()==True,"MINIMUM_PAYMENTS"] = data_df["MINIMUM_PAYMENTS"].mean()
```

```python
data_df["MINIMUM_PAYMENTS"].isnull().sum()
```

```python
data_df.loc[data_df["CREDIT_LIMIT"].isnull()==True] = data_df["CREDIT_LIMIT"].mean()
```

```python
data_df.duplicated().sum()
```

```python
data_df.drop("CUST_ID",axis=1,inplace=True)
```

```python
n =  len(data_df.columns)
n
```

```python
plt.figure(figsize=[10,50])

for i in range(n):
    plt.subplot(n,1,i+1)
    sns.distplot(data_df[data_df.columns[i]],kde_kws={"color":"b","lw":3,"label":"KDE"},hist_kws={"color":"r"})
    plt.title(data_df.columns[i])
plt.tight_layout()
    
```

```python
plt.figure(figsize=[10, 50])

for i in range(n):
    plt.subplot(n, 1, i+1)
    sns.kdeplot(data_df[data_df.columns[i]], color="b", lw=3, label="KDE",bw=0.1,cbar="b")
    plt.hist(data_df[data_df.columns[i]], color="r", alpha=0.5, label="hist")
    plt.title(data_df.columns[i])
    plt.legend()

plt.tight_layout()
```
